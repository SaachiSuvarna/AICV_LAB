```python
import cv2
from matplotlib import pyplot as plt
image=cv2.imread('cat.jfif')
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurred=cv2.GaussianBlur(gray,(5,5),0)

edged=cv2.Canny(blurred,5,40)
plt.imshow(image)
plt.show()
plt.imshow(blurred)
plt.show()
plt.imshow(edged)

wide_edged=cv2.Canny(blurred,10,100)
Mid_edged=cv2.Canny(blurred,30,150)
tight_edged=cv2.Canny(blurred,240,250)
plt.imshow(wide_edged)
plt.imshow(Mid_edged)
plt.imshow(tight_edged)
```

```python
contours, _=cv2.findContours(edged,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
image_copy=image.copy()
cv2.drawContours(image_copy,contours,-1,(0,255,0),2)
plt.imshow(edged)
plt.show()
plt.imshow(image_copy)
```
